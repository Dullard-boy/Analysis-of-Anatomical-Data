```python
from matplotlib import pyplot as plt
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
m = data['MATH']
e = data['ENGLISH']
r= data['RIGHT']
ax.scatter(m, r, color='y')
ax.scatter(e,r, color='g')
ax.set_xlabel('Intelligence Level')
ax.set_ylabel('Right Hand Index')
ax.set_title('Relationship between the Right Hand Index Finger and Intelligence')
ax.legend(labels=['MATH','ENGLISH'])
plt.savefig('Right', bbox_inches='tight')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
m = data['MATH']
e = data['ENGLISH']
r= data['LEFT']
ax.scatter(m, r)
ax.scatter(e,r, color='r')
ax.set_xlabel('Intelligence Level')
ax.set_ylabel('Left Hand Index')
ax.set_title('Relationship between the Left Hand Index Finger and Intelligence')
ax.legend(labels=['MATH','ENGLISH'])
plt.savefig('Left', bbox_inches='tight')
plt.show()
```
