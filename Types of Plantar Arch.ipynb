```python
import pandas as pd
```

```python
for i in data['Right PAI']:
    if i < 0.68:
        print('high arch')
    elif 0.68 < i < 0.91:
        print('normal arch')
    else:
        print('low arch')
```

```python
df =pd.cut(data['Right PAI'], bins=[0.16,0.68,0.91,2.43], labels=['high arch','normal arch','low arch'])
```

```python
df.value_counts()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
x = ['normal arch','high arch','low arch']
y= [116,63,58]
ax.set_xlabel('Type of Arch')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of Right Plantar Arch Types')
ax.bar(x,y, width=0.4, color='g')
plt.savefig('Right', bbox_inches='tight')
```
